```python
import tensorflow as tf
from tensorflow.keras import models,layers
# from tensorflow.keras.applications.inception_v3 import InceptionV3,preprocess_input
from keras.applications.vgg16 import VGG16,preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
```

```python
train_path = 'dog_cat/training_set'
valid_path = 'dog_cat/test_set'
```

```python
folders = glob('dog_cat/training_set/*')
folders
```

```python
IMAGE_SIZE = [224, 224]
```

```python
vgg = VGG16(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
vgg.input
```

```python
for layer in vgg.layers:
    layer.trainable = False
```

```python
folders = glob( r'C:\Users\Yuvraj Pardeshi\Downloads\dog_cat*')
print(len(folders))
```

```python
x = Flatten()(vgg.output)
prediction = Dense(len(folders), activation='softmax')(x)
model = models.Model(inputs=vgg.input, outputs=prediction)
model.summary()
```

```python
train_datagen = image.ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = image.ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory( r'C:\Users\Yuvraj Pardeshi\Downloads\dog_cat\train',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
test_set = train_datagen.flow_from_directory( r'C:\Users\Yuvraj Pardeshi\Downloads\dog_cat\test',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
r = model.fit_generator(
  training_set,
  validation_data=test_set,
  epochs=10,
  steps_per_epoch=len(training_set),
  validation_steps=len(test_set)
)
```

```python
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(r.history['accuracy'], label='train acc')
plt.plot(r.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```
